# Data Visualization

```
# loading matplotlib library
import matplotlib.pyplot as plt
```

```
x = [2,9,3,7]
y=[7,15,5,9]
plt.xlabel("Distance") 
plt.ylabel("Time")
plt.plot(x,y,label="Railway Track")
plt.scatter(x,y,s=200,label="Railway Station",marker="^",color="red")
plt.bar(x,y,label="Height of railway  station")
plt.grid(color="green")
plt.legend() # it will show label in the plot
plt.show()
```

```
# this will be another example of cricket data visualization
players = [
           'Dhoni',
           'Virat',
           'Rohit',
           'Sir Jadeja'
]
scores = [120,140,100,89]

# now plotting it
plt.xlabel("Players")
plt.ylabel("Scores")
# plt.bar(players,scores,label="2020 score of indian team")
plt.pie(scores,labels=players,explode=(0.2,0,0,0),shadow=True,autopct=("%1.1ff%%"))
# plt.grid(color="green")
# plt.legend()
plt.show()
```

# Web Scrapping

```
from urllib import request # for downloading data from url
from bs4 import BeautifulSoup # for scooping
# pointing to url
url = 'https://en.wikipedia.org/wiki/Machine_learning'
htmlData = request.urlopen(url)
# htmlData.read() # it will download data in html format
soupData = BeautifulSoup(htmlData,'html5lib')
#                  html data , html parser
# html parser -> is a collection of html tags that can scrap data from particular tag like h1,a or p
# selecting a particular tag for data scrape
all_p_tags = soupData.find_all('p')
myData = ""
for i in all_p_tags:
    myData += i.text

# print(myData)
```
