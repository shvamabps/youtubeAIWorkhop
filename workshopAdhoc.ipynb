# Web Scrapping

```
from urllib import request # for downloading data from url
from bs4 import BeautifulSoup # for scooping
import time
import re # import regular expression
# pointing to url
url = 'https://en.wikipedia.org/wiki/Machine_learning'
htmlData = request.urlopen(url)
# htmlData.read() # it will download data in html format
soupData = BeautifulSoup(htmlData,'html5lib')
#                  html data , html parser
# html parser -> is a collection of html tags that can scrap data from particular tag like h1,a or p
# selecting a particular tag for data scrape
all_p_tags = soupData.find_all('p')
myData = ""
for i in all_p_tags:
    myData += i.text
    time.sleep(1)

```
